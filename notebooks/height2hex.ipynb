```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd

# Example data
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
heights = np.array([10, 20, 30, 40, 50])  # Example height values

# Plot the data
plt.scatter(x, y, c=heights, cmap='viridis_r')
plt.colorbar()
plt.show()

# Normalize height values
norm = mcolors.Normalize(vmin=min(heights), vmax=max(heights))

# Choose a colormap
cmap = plt.get_cmap('viridis_r')

# Convert heights to hex colors
colors = [mcolors.to_hex(cmap(norm(h))) for h in heights]

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'color': colors})

# Save to CSV
csv_filename = 'points_with_colors.csv'
df.to_csv(csv_filename, index=False)

print(f"CSV file '{csv_filename}' created successfully!")

# Restore the data
df = pd.read_csv(csv_filename)

# Plot the data
plt.scatter(df['x'], df['y'], c=df['color'])
plt.colorbar()
plt.show()
```
